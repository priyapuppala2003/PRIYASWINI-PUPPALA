STEP 1: Upload & Load

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('/content/PROJECT DATASET.csv')
df.head()
```

1. Count Plot

```python
sns.countplot(x='Gender', data=df)
plt.title("Gender Distribution")
plt.show()
```

2. Bar Chart

```python
sns.barplot(x='Operating_System', y='App_Usage_Time', data=df)
plt.title("App Usage by Operating System")
plt.show()
```

3. Line Chart

```python
plt.plot(df['User_ID'], df['Screen_On_Time'])
plt.title("Screen Time Trend")
plt.xlabel("User ID")
plt.ylabel("Screen Time")
plt.show()
```

4. Histogram

```python
sns.histplot(df['Age'], bins=10, kde=True)
plt.title("Age Distribution")
plt.show()
```

5. Scatter Plot

```python
sns.scatterplot(x='App_Usage_Time', y='Battery_Drain', data=df)
plt.title("App Usage vs Battery Drain")
plt.show()
```

6. Box Plot

```python
sns.boxplot(x=df['Screen_On_Time'])
plt.title("Screen Time Boxplot")
plt.show()
```

7. Pie Chart

```python
df['Operating_System'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title("Operating System Distribution")
plt.ylabel("")
plt.show()
```

8. Pair Plot

```python
sns.pairplot(df)
plt.show()
```

9. Count Plot

```python
sns.countplot(x='User_Behavior_Class', data=df)
plt.title("User Behavior Class Distribution")
plt.show()
```
