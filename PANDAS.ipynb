**REMOVING DUPLICATES**

**Step 1: Import Pandas and Load Data**

```python
import pandas as pd

# Load your dataset (CSV or Excel)
df = pd.read_csv('/content/PROJECT DATASET.csv')  # Replace with your file path
df.head()  # Show first 5 rows
```

**Step 2: Check for Duplicates**

```python
# Check which rows are duplicates based on all columns
duplicates = df.duplicated()
print(df[duplicates])
```

```python
# Check duplicates based on a specific column, e.g., User_ID
duplicates = df.duplicated(subset='User_ID', keep=False)
print(df[duplicates])
```

**Step 3: Remove Duplicate Values**


```python
df = df.drop_duplicates(subset='User_ID', keep='first')
```

**Step 4: Verify Duplicates**

```python
print(df.duplicated().sum())
```

Step 5: Save Cleaned Data

```python
df.to_csv('cleaned_data.csv', index=False)
```

```python
df.to_excel('cleaned_data.xlsx', index=False)
```

**Data cleaning**

1. Import & Load Data

```python
import pandas as pd

df = pd.read_csv('/content/PROJECT DATASET.csv')
df.head()
```

2. Basic Data

```python
df.info()
```

```python
df.describe()
```

```python
df.shape
```

3.Handle Missing Values

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

4.Remove Duplicates

```python
df = df.drop_duplicates()
```

5. Rename Columns

```python
df.rename(columns={'old_name': 'new_name'}, inplace=True)
```

6.Handling missing values

```python
# Fill numeric columns
df.fillna(df.mean(numeric_only=True), inplace=True)

# Fill categorical columns
df.fillna('Unknown', inplace=True)
```

```python
df.columns = df.columns.str.strip().str.lower()
```

**DATA** PREPARATION


```python
df = pd.get_dummies(df, drop_first=True)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

df = pd.DataFrame(df_scaled, columns=df.columns)
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
