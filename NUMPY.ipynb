STEP 7: fULL CODE

```python
from google.colab import files
import pandas as pd
import numpy as np

# Assuming the file 'PROJECT DATASET.xlsx' has already been uploaded
# If not, uncomment the line below to upload it interactively:
# uploaded = files.upload()

# Load data from Excel using pandas
df = pd.read_excel('PROJECT DATASET.xlsx')

# Convert to NumPy array for consistency with other steps if needed,
# but we'll use the DataFrame for robust duplicate removal
data = df.to_numpy()

# Remove duplicates using pandas DataFrame for mixed data types
unique_data_df = df.drop_duplicates()
unique_data = unique_data_df.to_numpy()

# Check result
print("Original shape:", data.shape)
print("Cleaned shape:", unique_data.shape)

# Save cleaned data to CSV using pandas DataFrame
unique_data_df.to_csv('no_duplicates.csv', index=False)

# Download the cleaned file
files.download('no_duplicates.csv')
```

**DATA CLEANING**

1. Remove Missing Values

```python
# Convert to DataFrame to handle mixed types and NaN removal robustly
df_cleaned = pd.DataFrame(data).dropna()
data = df_cleaned.to_numpy()
```

2. Remove Duplicate Rows

```python
df_cleaned = pd.DataFrame(data).drop_duplicates()
data = df_cleaned.to_numpy()
```

**DATA PREPARATION**

3. Separate Features & Target

```python
X = data[:, :-1]   # all columns except last
y = data[:, -1]    # last column as target
```

4. Feature Scaling

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder

# Get column names for X from the original DataFrame (df) excluding the target column
x_column_names = df.columns[:-1]

# Convert X (NumPy array) back to a DataFrame to identify column types correctly
x_df = pd.DataFrame(X, columns=x_column_names)

# Identify numerical and categorical columns
numerical_cols = x_df.select_dtypes(include=['number']).columns
categorical_cols = x_df.select_dtypes(include=['object', 'category']).columns

# Create a preprocessor using ColumnTransformer
# Numerical columns will be scaled using StandardScaler
# Categorical columns will be one-hot encoded
preprocessor = ColumnTransformer(
    transformers=[
        ('num', StandardScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ],
    remainder='passthrough' # Keep any other columns if they exist (e.g., if there are columns that are neither num nor cat, which shouldn't be the case here)
)

# Apply the preprocessing to X
X_scaled = preprocessor.fit_transform(x_df)

print("Shape of X_scaled after preprocessing:", X_scaled.shape)
```

6. Train-Test Split

```python
# Set random seed
np.random.seed(42)

# Shuffle indices
indices = np.arange(X_scaled.shape[0])
np.random.shuffle(indices)

# Split 80-20
split = int(0.8 * len(indices))

train_idx = indices[:split]
test_idx = indices[split:]

X_train = X_scaled[train_idx]
X_test = X_scaled[test_idx]
y_train = y[train_idx]
y_test = y[test_idx]
```

STEP 7: Save Data
