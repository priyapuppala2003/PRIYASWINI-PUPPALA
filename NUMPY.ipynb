STEP 7: fULL CODE

```python
from google.colab import files
uploaded = files.upload()

import numpy as np

# Load CSV (numeric data only)
data = np.genfromtxt('PROJECT DATASET.csv', delimiter=',', skip_header=1)

# Remove duplicates
unique_data = np.unique(data, axis=0)

# Check result
print("Original:", data.shape)
print("Cleaned:", unique_data.shape)

# Save
np.savetxt('no_duplicates.csv', unique_data, delimiter=',')

# Download
files.download('no_duplicates.csv')
```

**DATA CLEANING**

1. Remove Missing Values

```python
# Remove rows with NaN values
data = data[~np.isnan(data).any(axis=1)]
```

2. Remove Duplicate Rows

```python
data = np.unique(data, axis=0)
```

**DATA PREPARATION**

3. Separate Features & Target

```python
X = data[:, :-1]   # all columns except last
y = data[:, -1]    # last column as target
```

4. Feature Scaling

```python
mean = np.mean(X, axis=0)
std = np.std(X, axis=0)

X_scaled = (X - mean) / std
```

6. Train-Test Split

```python
# Set random seed
np.random.seed(42)

# Shuffle indices
indices = np.arange(X_scaled.shape[0])
np.random.shuffle(indices)

# Split 80-20
split = int(0.8 * len(indices))

train_idx = indices[:split]
test_idx = indices[split:]

X_train = X_scaled[train_idx]
X_test = X_scaled[test_idx]
y_train = y[train_idx]
y_test = y[test_idx]
```

STEP 7: Save Data

```python
np.savetxt('X_train.csv', X_train, delimiter=',')
```

```python
np.savetxt('X_test.csv', X_test, delimiter=',')
```

```python
np.savetxt('y_train.csv', y_train, delimiter=',')
```

```python
np.savetxt('y_test.csv', y_test, delimiter=',')
```

STEP 8: Download The Files

```python
files.download('X_train.csv')
files.download('X_test.csv')
files.download('y_train.csv')
files.download('y_test.csv')
```
